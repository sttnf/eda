```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("./DATA/data_bersih_drop.xlsx")

df.index = df.index + 1

df.info()

# df.head()
```

```python
correlation = df["Usia"].corr(df["Pendapatan"])
print(f"Korelasi antara Usia dan Pendapatan adalah {correlation}")

r2 = correlation ** 2
print(f"Koefisien determinasi (R²) antara Usia dan Pendapatan adalah {r2}")
```

```python
df["Pendapatan"] = df["Pendapatan"] / 1_000_000
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["Usia"], df["Pendapatan"], alpha=0.7)
plt.title("Scatter Plot Usia vs Pendapatan")
plt.xlabel("Usia / Umur")
plt.ylabel("Pendapatan (Juta Rupiah)")

plt.grid(True)
plt.show()
```
